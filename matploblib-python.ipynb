```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list = [20,25,30,35,40,45,50,55,60]
```

```python
weight_list = [70,75,73,85,89,90,75,84,90]
```

```python
age_list_numpy = np.array(age_list)
weight_list_numpy = np.array(weight_list)
```

```python
age_list_numpy
```

```python
weight_list_numpy
```

```python
plt.plot(age_list_numpy,weight_list_numpy, "g")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age - Weight Graph")
plt.show()
```

```python
my_numpy1 = np.linspace(0,15,20)
my_numpy1
```

```python
my_numpy2 = my_numpy1 ** 3
my_numpy2
```

```python
plt.plot(my_numpy1,my_numpy2,"y-*")
plt.show()
```

```python
plt.subplot(1,2,1) # 1 Row, 2 Column, 1. Graph
plt.plot(my_numpy1,my_numpy2,"r--")

plt.subplot(1,2,2) # 1 Row, 2 Column, 2. Graph
plt.plot(my_numpy2,my_numpy1,"g*")
plt.show()
```

```python
my_figure = plt.figure()

my_axes = my_figure.add_axes([0.1,0.1,0.4,0.4])
my_axes.plot(my_numpy1, my_numpy2,"g")
my_axes.set_xlabel("X Data")
my_axes.set_ylabel("Y Data")
my_axes.set_title("Graph")
plt.show()
```

```python
my_figure2 = plt.figure()

my_axes2 = my_figure2.add_axes([0.1,0.1,0.2,0.2])
my_axes2.plot(my_numpy1, my_numpy2,"g*")
my_axes2.set_xlabel("X Data Small")
my_axes2.set_ylabel("Y Data Small")
my_axes2.set_title("Small Graph")

my_axes3 = my_figure2.add_axes([0.3,0.4,0.4,0.4])
my_axes3.plot(my_numpy1, my_numpy2,"r*-")
my_axes3.set_xlabel("X Data Large")
my_axes3.set_ylabel("Y Data Large")
my_axes3.set_title("Large Graph")

plt.show()
```

```python
my_numpy3 = np.linspace(0,15,20)
my_numpy4 = my_numpy3 + 3
```

```python
# (my_fig, my_axe) = plt.subplots(2,2)
(my_fig, my_axe) = plt.subplots()

my_axe.plot(my_numpy3,my_numpy4, color="#CD621D", alpha=1)
my_axe.plot(my_numpy4,my_numpy3, color="#B81DC1", alpha=0.2)
plt.show()
```

```python
(new_fig, new_axe) = plt.subplots()
new_axe.plot(my_numpy3, my_numpy3 + 2, color="#B81DC1", linewidth=2.4)
new_axe.plot(my_numpy3, my_numpy3 + 3, color="blue", linestyle="-.")
new_axe.plot(my_numpy3, my_numpy3 + 4, color="black", linestyle="--", marker="o", markersize=8)
plt.show()
```

```python
plt.scatter(my_numpy3, my_numpy4)
plt.show()
```
